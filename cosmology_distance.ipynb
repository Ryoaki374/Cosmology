```python
# module
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt

k = 1.380649e-23
h = 6.62607015e-34
T = 2.725
Lambda = 0.7
coef = 3e+3

# function
def f(x):
  return 1/(np.sqrt(0.3/0.7*(1+x)**3 + 1))

# upper,lower
upper = 10
lower = 0


ans,err = integrate.quad(f,lower,upper)

print(1/np.sqrt(Lambda)*3e+3*ans)
print(err)
```

```python
num_z = []
com_dist = []
ang_dist = []
lum_dist = []
```

```python
for i in range(1,1000,1):
    z = i/100                       #z = 0.01 ~ 10
    ans,err = integrate.quad(f,0,z) #scipi integrate (Gaussian quadrature) https://en.wikipedia.org/wiki/Gaussian_quadrature
    com_calc = 1/np.sqrt(Lambda)*3e+3*ans
    num_z.append(z)                 #red shift
    com_dist.append(com_calc)       #comving distance
    ang_dist.append(com_calc/(1+z)) #angular distance
    lum_dist.append(com_calc*(1+z)) #luminocity distance
```

```python
print(com_dist)
```

```python
plt.plot(num_z,com_dist,label = r'$\chi$',color = 'black')
plt.plot(num_z,ang_dist,label = r'$d_A$',color = 'r')
plt.plot(num_z,lum_dist,label = r'$d_L$',color = 'g')
plt.xscale('log')
plt.xlabel('Red shift  '+r'$z$')
plt.yscale('log')
plt.ylabel('Distance  ' +r'$d$'+ '  /  ' + r'$h^{-1}$' + 'Mpc')
plt.xlim(0.05,10)
plt.ylim(100,12000)
plt.legend()
```
