**目標**: $\mathcal H(x),\eta(x)$ と密度分率の進化、$z_{\rm eq}$。所要時間: 5分。

```
import numpy as np, matplotlib.pyplot as plt
from pathlib import Path
from cmbcore.plotstyle import use_style; use_style()
np.random.seed(5220)
FIG = Path('..') / 'figures'
from cmbcore import Params, BackgroundCosmology
p = Params.fiducial(); bg = BackgroundCosmology(p)
print('z_eq = %.0f, age = %.2f Gyr, eta0 = %.0f Mpc' % (bg.z_eq(), bg.t0/3.1557e16, bg.eta0/3.0857e22))
```

## F1.1 密度分率 $\Omega_i(x)$

```
x = np.linspace(-15,0,400); Om = bg.Omega(x)
for key in ['gamma','nu','b','c','Lambda']: plt.plot(x,Om[key],label=key)
plt.axvline(np.log(1/(1+bg.z_eq())),ls='--',c='k',label='z_eq')
plt.xlabel('x=ln a'); plt.ylabel('Omega_i'); plt.legend(); plt.show()
```

## F1.2 共形ハッブル $\mathcal H(x)/H_0$ と漸近則

```
plt.semilogy(x, bg.Hp(x)/p.H0, label='Hp/H0')
plt.semilogy(x, np.exp(-x), '--', label='~e^{-x} (rad)')
plt.semilogy(x, np.exp(-x/2), ':', label='~e^{-x/2} (mat)')
plt.xlabel('x'); plt.legend(); plt.ylim(0.5,1e3); plt.show()
```
