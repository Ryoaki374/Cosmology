```
import numpy as np, matplotlib.pyplot as plt
from cmbcore import Params, BackgroundCosmology
from cmbcore.plotstyle import use_style; use_style()
p = Params.fiducial(); bg = BackgroundCosmology(p)
```

## 密度分率の進化

```
x = np.linspace(-15, 0, 400)
Om = bg.Omega(x)
for key in ['gamma','nu','b','c','Lambda']:
    plt.plot(x, Om[key], label=key)
plt.axvline(np.log(1/(1+bg.z_eq())), ls='--', c='k')
plt.xlabel('x = ln a'); plt.ylabel('Omega_i(x)')
plt.legend(); plt.title('density fractions'); plt.show()
```
