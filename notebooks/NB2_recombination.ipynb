# NB2 — 再結合（第3章）

**目標**: Saha vs Peebles の $X_e$、$\tau,\tilde g$、$z_*$。所要時間: 5分。

```
import numpy as np, matplotlib.pyplot as plt
from pathlib import Path
from cmbcore.plotstyle import use_style; use_style()
np.random.seed(5220)
FIG = Path('..') / 'figures'
from cmbcore import Params, BackgroundCosmology, Recombination
p = Params.fiducial(); bg = BackgroundCosmology(p); rec = Recombination(bg,p)
print('z_* = %.0f, r_s(z_*) = %.1f Mpc, int g dx = %.5f' % (rec.z_star(), rec.r_s(rec.x_star())/3.0857e22, rec.visibility_norm()))
```

## F2.1 $X_e(x)$（Saha 単独 vs Peebles）

```
x = np.linspace(-9,-5,500); z = np.exp(-x)-1
plt.semilogy(z, rec.Xe(x), label='Saha+Peebles')
plt.semilogy(z, rec._saha_Xe(x), '--', label='Saha only')
plt.axvline(rec.z_star(), c='r', ls=':', label='z_*')
plt.gca().invert_xaxis(); plt.xlabel('z'); plt.ylabel('X_e'); plt.legend(); plt.show()
```

## F2.3 可視度関数 $\tilde g(x)$

```
plt.plot(x, rec.g_tilde(x)); plt.xlabel('x=ln a'); plt.ylabel('g~(x)')
plt.title('visibility (peaks at last scattering)'); plt.show()
```

## 追加実験: 再電離 ON での第2の山（第13章の伏線）

```
pr = Params(tau_reio=1.0, z_reio=8.0)
rec2 = Recombination(BackgroundCosmology(pr), pr)
xx = np.linspace(-12,0,800)
plt.plot(xx, rec.g_tilde(xx), label='no reion')
plt.plot(xx, rec2.g_tilde(xx), label='reion z_re=8')
plt.yscale('log'); plt.ylim(1e-4,None); plt.xlabel('x'); plt.legend(); plt.show()
```
