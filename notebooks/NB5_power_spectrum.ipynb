**目標**: $\mathcal D_\ell^{TT}$、ピーク構造の読み取り。所要時間: <1分（キャッシュ利用）。

```
import numpy as np, matplotlib.pyplot as plt
from pathlib import Path
from cmbcore.plotstyle import use_style; use_style()
np.random.seed(5220)
FIG = Path('..') / 'figures'
d = np.load(FIG/'cls_fiducial.npz')
ells, Dl = d['ells'], d['cls']
print('first peak at l =', int(ells[np.argmax(np.where(ells>=150,Dl,0))]))
```

## F5.1 $\mathcal D_\ell^{TT}$（fiducial）と Planck 2018 binned TT

```
plt.plot(ells, Dl, label='cmbcore')
pl = Path('data')/'planck2018_tt_binned.csv'
if pl.exists():
    P = np.genfromtxt(pl, delimiter=',', comments='#')  # cols: l, Dl, err
    plt.errorbar(P[:,0], P[:,1], yerr=P[:,2], fmt='.', ms=6,
                 label='Planck 2018 (illustrative)')
plt.xlabel('multipole l'); plt.ylabel('D_l [uK^2]'); plt.legend(); plt.show()
```

## F5.4 ピーク検出表

```
from scipy.signal import find_peaks
lf=np.arange(120,1500); Df=np.interp(lf,ells,Dl)
idx,_=find_peaks(Df, prominence=0.1*Df.max(), distance=120)
for j,l in enumerate(lf[idx][:4]): print('peak %d: l=%d, D=%.0f uK^2'%(j+1,l,Df[idx][j]))
```
