# NB3 — 単一モードの摂動進化（第6–9章; スライド再現）

**目標**: $\Theta_0,\delta_b,\delta_c,\Phi,\Psi$ の進化とスケール依存。所要時間: 1分。

```
import numpy as np, matplotlib.pyplot as plt
from pathlib import Path
from cmbcore.plotstyle import use_style; use_style()
np.random.seed(5220)
FIG = Path('..') / 'figures'
from cmbcore import Params, BackgroundCosmology, Recombination
from cmbcore.perturbations import PerturbationSolver
from cmbcore import constants as const
p=Params.fiducial(); bg=BackgroundCosmology(p); rec=Recombination(bg,p)
solver=PerturbationSolver(bg,rec,p)
H0_c = p.H0/const.c   # 1/m
ks = {n: n*H0_c for n in (10,100,1000)}
res = {n: solver.solve(k) for n,k in ks.items()}
```

## F3.1/F3.6 光子モノポール・多重極 $\Theta_\ell(a)$, $k=1000H_0/c$

```
r=res[1000]; x=np.linspace(-12,0,500); a=np.exp(x)
for l in (0,1,2): plt.plot(a, r[f'Theta{l}'](x), label=f'Theta_{l}')
plt.xscale('log'); plt.xlabel('a'); plt.legend(); plt.title('acoustic oscillations'); plt.show()
```

## F3.5 CDM 密度 $\delta_c(a)$: スケール別の成長開始

```
for n in (10,100,1000):
    plt.loglog(a, np.abs(res[n]['delta_c'](x)), label=f'k={n}H0/c')
plt.xlabel('a'); plt.ylabel('|delta_c|'); plt.legend(); plt.show()
```

## F3.7 重力ポテンシャル $\Phi,\Psi$（地平線進入・等密度・$\Lambda$ 期の減衰）

```
for n in (10,1000):
    plt.plot(a, res[n]['Phi'](x), label=f'Phi k={n}')
plt.xscale('log'); plt.xlabel('a'); plt.ylabel('Phi'); plt.legend(); plt.show()
```

## 読み取り: 大スケールは超地平線で凍結、小スケールは音響振動＋減衰。
