# NB4 — 転送関数とソース関数（第10章）

**目標**: ソース $\tilde S(k,x)$ の4成分、視線積分 $\Theta_\ell(k)$。所要時間: 2分。

```
import numpy as np, matplotlib.pyplot as plt
from pathlib import Path
from cmbcore.plotstyle import use_style; use_style()
np.random.seed(5220)
FIG = Path('..') / 'figures'
from cmbcore import Params, PowerSpectrum
from cmbcore.spectrum import k_grid
from cmbcore import constants as const
ps = PowerSpectrum(Params.fiducial())
ks = k_grid(nk=80)/const.Mpc
ells = np.array([2,100,220,500,1000])
kf, Th = ps.transfer(ells, ks_si=ks)
```

## F4.2 転送関数 $\Theta_\ell(k)$（細kグリッドで滑らか）

```
kmpc = kf*const.Mpc
for i,l in enumerate(ells): plt.plot(kmpc, Th[i], label=f'l={l}')
plt.xlabel('k [1/Mpc]'); plt.ylabel('Theta_l(k)'); plt.xlim(0,0.1); plt.legend(); plt.show()
```

## F4.4 ベッセルカーネル $j_\ell(k\chi)$ が角度射影を担う
$\Theta_\ell(k)=\int \tilde S(k,x)\,j_\ell[k(\eta_0-\eta)]\,dx$。

```
from scipy.special import spherical_jn
chi0 = ps.bg.eta0
kk = np.linspace(0,0.1,400)
for l in (100,220,500): plt.plot(kk, spherical_jn(l, kk*chi0/const.Mpc*const.Mpc), label=f'l={l}')
plt.xlabel('k [1/Mpc]'); plt.ylabel('j_l(k chi0)'); plt.legend(); plt.show()
```

## 検収: $\Theta_\ell(k)$ はグリッド起因のギザつきがない。
## 【課題】 ソースを `ps.sources(ks)` で取り出し4成分を分解せよ。
