# NB7 — 解析近似 vs 数値（第8–9,12章）

**目標**: 音地平線・Silk スケール・ピーク位置の解析予言を数値と突き合わせる。ここでは教育目的で解析式を直接書く。所要時間: <1分。

```
import numpy as np, matplotlib.pyplot as plt
from pathlib import Path
from cmbcore.plotstyle import use_style; use_style()
np.random.seed(5220)
FIG = Path('..') / 'figures'
from cmbcore import Params, BackgroundCosmology, Recombination
from cmbcore import analytic, constants as const
p=Params.fiducial(); bg=BackgroundCosmology(p); rec=Recombination(bg,p)
lA = analytic.acoustic_scale(bg,rec)
kD = analytic.silk_scale(bg,rec)*const.Mpc
print('l_A = %.1f, k_D = %.4f /Mpc, 100 theta_* = %.4f' % (lA, kD, analytic.theta_star(bg,rec)))
```

## F7.4 ピーク位置予言 $\ell_m\approx\ell_A(m-\phi)$ vs 数値検出

```
pk = analytic.peak_positions(bg,rec,n=4)
d = np.load(FIG/'cls_fiducial.npz'); ells,Dl=d['ells'],d['cls']
from scipy.signal import find_peaks
lf=np.arange(120,1500); Df=np.interp(lf,ells,Dl)
idx,_=find_peaks(Df,prominence=0.1*Df.max(),distance=120); num=lf[idx][:4]
for m,(a_,n_) in enumerate(zip(pk,num),1): print('peak %d: analytic %.0f, numeric %d'%(m,a_,n_))
```
