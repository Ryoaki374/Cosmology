```python
# module
from scipy import integrate
import numpy as np
import matplotlib.pyplot as plt

k = 1.380649e-23 
h = 6.62607015e-34
T = 2.725

# function
def f(x):
  return x/(np.exp(x)-1)

# upper,lower
upper = 150e+9*0.85*h/k/T
lower = 150e+9*1.15*h/k/T

ans,err = integrate.quad(f,upper,lower)

coef = (k*T)**2/h
print(coef*ans)
print(coef*err)
```

```python
a = 150*0.85*1e9
b = 150*1.15*1e9
m = 100
n = int(m/2)

# 積分する関数の定義
def f(x):
  return x/(np.exp(x)-1)

def Simpson(f,a,b,m):
    df = (b-a)/(2*n)
    h = (b-a)/m
    sum = 0
    for i in range(1,n+1):
        x1 = a + df*(2*i -2)
        x2 = a + df*(2*i -1)
        x3 = a + df*(2*i)
        f1 = f(x1)
        f2 = f(x2)
        f3 = f(x3)
        sum += h*(f1 + 4*f2 + f3)/3
    return sum

# 結果の表示
print(Simpson(f,a,b,m))
```

```python

no = 3.06
ne = 3.4

def f_no(x):
    return np.sin(x)/no

def f_ne(x):
    return np.sin(x)/ne

def e_no(x):
    return np.sin(x)*no

def e_ne(x):
    return np.sin(x)*no
```

```python
arcmin = 1/60.
print(arcmin*np.pi/180.)
```

```python
theta_w = []
output_no = []
output_ne = []



for i in range (50):
    print(arcmin*i*np.pi/180.)
    output_no.append(np.arcsin(f_no(arcmin*i*np.pi/180.)))
    output_ne.append(np.arcsin(f_ne(arcmin*i*np.pi/180.)))
    theta_w.append(arcmin*i*np.pi/180.)


```

```python
print(output_no)
print(len(theta_w),len(output_no))
```

```python
plt.rcParams["font.size"] = 22

fig,ax = plt.subplots(figsize=(20,12))
#ax_r = ax.twinx()
ax.plot(theta_w,output_no,color = 'red',label = 'no')
ax.plot(theta_w,output_ne,color = 'blue',label = 'ne')
#ax.scatter(df_VNA['freq_round'],df_VNA['tr_VNA_ave'],color = 'blue',label = l1)
ax.set_xlabel("theta_w")
ax.set_ylabel('theta_out')
ax.tick_params(direction = "in")
ax.legend()
plt.show()
```
